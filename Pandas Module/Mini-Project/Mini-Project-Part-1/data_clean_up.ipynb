```python
# Dependencies
import pandas as pd
```

```python
# load CSV
bootcamp_file = "Resources/2016-FCC-New-Coders-Survey-Data.csv"

```

```python
# Read with pandas
bootcamp_df=pd.read_csv(bootcamp_file)
```

```python
# Inspect all columns
bootcamp_df.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Use iloc to accomplish this (remember that you need to take the position of the rows and columns into account)

bootcamp_df_new = bootcamp_df.iloc[:,[0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
bootcamp_df_new
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
# Hint use the df.replace functio
bootcamp_df_new['AttendedBootcamp']= bootcamp_df_new['AttendedBootcamp'].replace({1.0:'Yes', 0.0:'No' })

bootcamp_df_new
```
