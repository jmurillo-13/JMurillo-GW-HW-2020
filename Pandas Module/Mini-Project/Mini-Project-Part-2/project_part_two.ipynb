```python
# Dependencies
import pandas as pd
```

```python
# load CSV
new_coder_data = pd.read_csv("Resources/2016-FCC-New-Coders-Survey-Data.csv", low_memory=False)
new_coder_data.head()
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
# You may need to encode this file

```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Consider using iloc
new_coder_data_new = new_coder_data.iloc[:,[0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
new_coder_data_new

```

```python
# Change reading 0 and 1 to No and Yes, respectively
new_coder_data_clean= new_coder_data_new.replace({1.0:'Yes', 0.0:'No' })


new_coder_data_clean.head()
```

```python
# Extract rows for only those who attended a bootcamp

attended_bc_df = new_coder_data_clean.loc[new_coder_data_clean['AttendedBootcamp'] == "Yes",:]
attended_bc_df
```

```python
# Create DataFrame of the various boot camps along with the number of participants (enrollees)
# Create a DataFrame from using value_counts

bootcamp_attendees= pd.DataFrame(attended_bc_df['BootcampName'].value_counts())
bootcamp_attendees.reset_index(inplace=True)
bootcamp_attendees = bootcamp_attendees.rename(columns = {'index': 'BootcampName',
                                     'BootcampName': "Number of Attendees"})


bootcamp_attendees
#bootcamp_attendees = bootcamp_attendees.reset_index

```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
reccomenders_df = attended_bc_df.loc[attended_bc_df['BootcampRecommend']=="Yes",:]
bootcamps = reccomenders_df.groupby('BootcampName')

reccomender_summary_df = bootcamps['BootcampRecommend'].count().to_frame()
reccomender_summary_df.reset_index(inplace=True)


reccomender_summary_df
#number_recommenders = bootcamps
```

```python
# Merge the two created data frames on the name of tbe bootcamp
#merge_df = pd.merge(reccomender_summary_df, bootcamp_attendees, left_index=True, right_index = True)

merge_df = pd.merge(reccomender_summary_df, bootcamp_attendees, on ='BootcampName')
merge_df = merge_df.rename(columns = {'BootcampName': 'Bootcamp Name'})

merge_df
```

```python
# Calculate percentage of each bootcamp's students who are recommenders
# Think of what should be the numerator and the demoninator

merge_df['Percent Reccomend'] = (merge_df['BootcampRecommend']/merge_df['Number of Attendees'])*100
merge_df = merge_df.sort_values(['Percent Reccomend'], ascending=False)


# Sort results in descending order
merge_df = merge_df.sort_values(['Percent Reccomend'], ascending=False)

# Format for percentages
merge_df['Percent Reccomend'] = merge_df['Percent Reccomend'].map("{:.2f}%".format)
merge_df
```

```python
# Export to excel and remove index
merge_df.to_excel("output/mini_assignment_2.xlsx", index=False)
```
